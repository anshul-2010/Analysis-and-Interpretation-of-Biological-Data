## Question 7
### Are the variances between chlorophyll a and chlorophyll b measurements different significantly? Perform appropriate statistical tests to support your claim. Compare variances of chlorophyll content from Deciduous forests separately, Evergreen forests separately and both the forests together. (12 marks)

```python
# Null Hypothesis: Ho = σ2(c_a) = σ2(c_b) for deciduous forests
# Alternate Hypothesis: H1 = σ2(c_a) != σ2(c_b) for deciduous forests
# n1 = 50; n2 = 50
# v1 = 49; v2 = 49
# α = 5% = 0.05
var_a = deciduous.Chlorophyll_a.var()
var_b = deciduous.Chlorophyll_b.var()
f_val_d = var_a / var_b
p_val_d = 1 - stats.f.cdf(f_val_d, 49, 49)
print("p value deciduous:",p_val_d)
# Since p value is less than α, I am rejecting null hypothesis
"""H1 = σ2(c_a) != σ2(c_b) for deciduous forests"""

# Null Hypothesis: Ho = σ2(c_a) = σ2(c_b) for evergreen forests
# Alternate Hypothesis: H1 = σ2(c_a) != σ2(c_b) for evergreen forests
# n1 = 50; n2 = 50
# v1 = 49; v2 = 49
# α = 5% = 0.05
var_a = evergreen.Chlorophyll_a.var()
var_b = evergreen.Chlorophyll_b.var()
f_val_e = var_a / var_b
p_val_e = 1 - stats.f.cdf(f_val_e, 49, 49)
print("p value evergreen:",p_val_e)
# Since p value is less than α, I am rejecting null hypothesis
"""H1 = σ2(c_a) != σ2(c_b) for evergreen forests"""

# Null Hypothesis: Ho = σ2(c_a) = σ2(c_b) for all forests
# Alternate Hypothesis: H1 = σ2(c_a) != σ2(c_b) for all forests
# n1 = 100; n2 = 100
# v1 = 99; v2 = 99
# α = 5% = 0.05
var_a = data.Chlorophyll_a.var()
var_b = data.Chlorophyll_b.var()
f_val_a = var_a / var_b
p_val_a = 1 - stats.f.cdf(f_val_a, 99, 99)
print("p value all forests:",p_val_a)
# Since p value is less than α, I am rejecting null hypothesis
"""H1 = σ2(c_a) != σ2(c_b) for all forests together"""
print()
```

## Question 8
### Finally, test whether the mean of chlorophyll a is greater than mean of chlorophyll b using appropriate statistical test in all three combinations: Deciduous forests separately, Evergreen forests separately and both the forests together. Clearly state your null hypothesis, chosen significance criteria and the result of hypothesis testing. (12 marks)

```python
"""Since the null hypothesis was rejected and conclusion obtained was that the variances are not equal,
   I will use the separate variance t-test to compare the means. I am considering alpha = 0.05 as significance level"""
print(end="")
```

```python
# Deciduous forests
stat_val_d, p_val_d = stats.ttest_ind(deciduous.Chlorophyll_a, deciduous.Chlorophyll_b, equal_var = False, alternative = "greater")
print("Statistic value of deciduous forests ttest:", stat_val_d)
print("p value of deciduous forests ttest:", p_val_d)

# Evergreen forests
stat_val_e, p_val_e = stats.ttest_ind(evergreen.Chlorophyll_a, evergreen.Chlorophyll_b, equal_var = False, alternative = "greater")
print("Statistic value of evergreen forests ttest:", stat_val_e)
print("p value of evergreen forests ttest:", p_val_e)

# Both forests together
stat_val_a, p_val_a = stats.ttest_ind(data.Chlorophyll_a, data.Chlorophyll_b, equal_var = False, alternative = "greater")
print("Statistic value of all forests ttest:", stat_val_a)
print("p value of all forests ttest:", p_val_a)
```

```python
"""p value in each of the above three tests is less than alpha, which implies that we are rejecting the null hypothesis
   and accepting the alternate hypothesis. The alternate hypothesis says:
   mean of chlorophyll_a is greater than mean of chlorophyll_b. This is true for deciduous forests only, evergreen forests only,
   and both forests together"""
print(end="")
```
